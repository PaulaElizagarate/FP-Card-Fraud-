# Client Clustering 

```python
import pandas as pd
import numpy as np
```

### Dataset

```python
df = pd.read_csv("../data/clients_cluster_def.csv")
df.head()
```

```python
df= df.drop("Unnamed: 0",axis= 1)
```

```python
df1 = pd.read_csv("../data/clusters.csv")
df1.head()
```

```python
df1= df1.drop("Unnamed: 0",axis= 1)
```

```python
clusters= pd.concat([df, df1], axis=1)
clusters.head()
```

```python
# saving our dataframe 
clusters.to_csv('../data/clusters_tableau.csv')
```
